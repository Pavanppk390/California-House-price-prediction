# EDA -> EXPLORATORY DATA ANALYSIS









Boxplot-> to check the outliers

```python
fig,ax= plt.subplots(figsize=(15,15))
ax= sns.boxplot(df)
plt.show()
```

**split** the data into dependent and independent  features

```python
x= df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
X_train,x_test,Y_train,y_test = train_test_split(x,y,random_state=96)
```

We keep longitude and latitude because they offer different sets of information as longitude represents east west  while latitude represents north south

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

vif = pd.DataFrame()
vif["Feature"] = X_train.columns
vif["VIF"] = [
    variance_inflation_factor(X_train.values, i)
    for i in range(X_train.shape[1])
]

print(vif.sort_values("VIF", ascending=False))
```

```python
X_train_new = X_train.drop(columns=['AveBedrms'])
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

vif = pd.DataFrame()
vif["Feature"] = X_train_new.columns
vif["VIF"] = [
    variance_inflation_factor(X_train_new.values, i)
    for i in range(X_train_new.shape[1])
]

print(vif.sort_values("VIF", ascending=False))
```

THE high vif is reduced for AveRooms

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_norm = scaler.fit_transform(X_train_new)
```

```python
fig,ax= plt.subplots(figsize=(15,15))
sns.boxplot(X_train_norm,ax=ax)
plt.savefig("normalized_training_set_boxplot.jpg")
plt.show()
```

```python
x_test_new = x_test.drop(columns="AveBedrms")
```

```python
X_test_norm = scaler.transform(x_test_new)
```

```python
print(X_train_new.shape)
print(x_test_new.shape)
print(X_train_norm.shape)
print(X_test_norm.shape)
```

```python
from sklearn.linear_model import LinearRegression
Regression = LinearRegression()
Regression.fit(X_train_norm,Y_train)
```

```python
print(Regression.coef_)
```

```python
print(Regression.intercept_)
```

# **MODEL PREDICTION**

```python
reg_pred = Regression.predict(X_test_norm)
```

```python
reg_pred
```

```python
error = y_test-reg_pred
error
```

Checking the normality of the errors or residuals


```python
sns.displot(error,kde= True)
plt.savefig("error_dist.jpg")
plt.show()
```

Model performance

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(r2_score(y_test,reg_pred))
```

Saving the model

```python
import pickle
pickle.dump(Regression,open('model.pkl',"wb"))
```
